```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tmp = plt.imread('movies/snow/snow_0001.tif')
tmp.shape
```

```python
def read_series(fmask,first,last) :
    tmp = plt.imread(fmask.format(first)).mean(axis=2)
    
    img=np.zeros([last-first+1,tmp.shape[0],tmp.shape[1]])
    img[0]=tmp
    
    for idx in range(first+1,last+1) :
        img[idx-first]=plt.imread(fmask.format(idx)).mean(axis=2)
        
    return img
```

```python
img = read_series('movies/snow/snow_{0:04d}.tif',first=1, last=300)
```

```python
th=100
plt.imshow(np.max(th<img[0:40],axis=0),interpolation='none')
```
